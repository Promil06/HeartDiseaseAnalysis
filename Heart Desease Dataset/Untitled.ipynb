## Analysis of Categorical columns

```python
df.groupby('Sex').size().plot(kind = 'pie', ylabel = 'Sex ratio', figsize = (3,4), autopct = '%0.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', ylabel = 'Chest Pain Type', figsize = (3,4), autopct = '%0.1f')
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', figsize = (3,4), autopct = '%0.1f')
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', ylabel = 'ST Slope', figsize = (3,4), autopct = '%0.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', ylabel = 'Heart Disease Patients', figsize = (3,4), autopct = '%0.1f')
```
